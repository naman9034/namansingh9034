```python
 #Step-1  Import required modules

import numpy as np
import pandas as pd

# Sklearn library
from sklearn import preprocessing
```

```python
# Step -2  Read data file

df1 = pd.read_csv('https://github.com/naman9034/admission-pre/raw/naman9034-patch-2/Data_for_preprocessing1.csv')
df1
```

```python
# Step -3 Assign input

X = df1.iloc[:, :-1].values
#This function return all columns except last as an array
```

```python
X
```

```python
# we can write above statement as column start from 0

a=df1.iloc[:, :3].values
a
```

```python
#Step 4 - Manage missing values
#fill missing value mean value

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan , strategy = 'mean')
```

```python
#Now decide actual column for treatment of missing value

imputer = imputer.fit(X[:,1:3])
```

```python
X
```

```python
#Here actual operation begins
X[:, 1:3] = imputer.transform(X[:, 1:3])
X
# Now numeric column with Nan value filled with mean
```

```python
# We can use single function in place of fit and transform as fit_transform

X[:, 1:3] = imputer.fit_transform(X[:, 1:3])
X
```

```python
df1
```

```python
#Step 5 - Now treat categorical data

from sklearn.preprocessing import LabelEncoder
#use this class to convert country (text) value in numeric value
labelencoder_X = LabelEncoder()

labelencoder_X.fit(X[:,0])
X[:,0] = labelencoder_X.transform(X[:,0])
X
```

```python
# fit_transform treat categorical data in single step

X[:,0] = labelencoder_X.fit_transform(X[:,0])
X
```
